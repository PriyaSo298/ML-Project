```python
# import Library
import pandas as pd
import numpy as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Algorith
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier

#Evaluation Library/Packages
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,roc_curve
```

```python
# Load Iris Dataset from seaborn library
df=sns.load_dataset('iris')
```

# Univarient Analysis
##### Distribution of one perticular column

```python
def Distribution(x):
    for i in x:
        plt.figure(figsize=(5,5))
        sns.displot(df[i],kind="kde")
Distribution(df.columns)
```

# Bivarient analysis
##### Correlation between two columns

```python
import warnings
warnings.filterwarnings("ignore")
def correlaton(x):
    for i in x:
        for j in x:
            if(i!=j):
                plt.figure(figsize=(5,5))
                sns.scatterplot(df[i],df[j])
correlaton(df.columns)
```

# Multinomial Analysis
##### corelation between multiple column

```python
sns.heatmap(df.corr())
```

```python
# Label encoding on depending column
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
df["species"]=LE.fit_transform(df["species"])
```

```python
#split data in training and testing part for preventing data leakage
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,0:4],df.iloc[:,4:],test_size=0.20,random_state=11)
```

# Model builder

```python
Logi_reg=LogisticRegression()
Logi_reg.fit(x_train,y_train)
y_predict=Logi_reg.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_predict)
accuracy
```

```python
#Classification Report
class_re=classification_report(y_test,y_predict)
print(class_re)
```

```python
#Plotting Confusion matrix
confu=confusion_matrix(y_test,y_predict)
print(confu)
print("Confusion matrix is ")
from sklearn import metrics
metrics.plot_confusion_matrix(Logi_reg,x_test, y_test)
plt.show()
```

# KNN

```python

KNN=KNeighborsClassifier()
KNN.fit(x_train,y_train)
y_pred=KNN.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)
print(accuracy)

#Classification Report
class_re=classification_report(y_test,y_pred)
print(class_re)

#Plotting Confusion matrix
confu=confusion_matrix(y_test,y_pred)
print(confu)
print("Confusion matrix is ")
from sklearn import metrics
metrics.plot_confusion_matrix(KNN,x_test, y_test)
plt.show()
```

# Decision Tree

```python

DT=DecisionTreeClassifier()
```

```python
DT.fit(x_train,y_train)
y_predicts=DT.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_predicts)
print(accuracy)

#Classification Report
class_re=classification_report(y_test,y_predicts)
print(class_re)

#Plotting Confusion matrix
confu=confusion_matrix(y_test,y_predicts)
print(confu)
print("Confusion matrix is ")
from sklearn import metrics
metrics.plot_confusion_matrix(DT,x_test, y_test)
plt.show()
```
