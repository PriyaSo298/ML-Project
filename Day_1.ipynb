```python
x=data.iloc[:,1:]
```

```python
y=data[["Survived"]]
```

```python
from sklearn.preprocessing import StandardScaler
a=StandardScaler()
x_scaler=a.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x_scaler,y,test_size=0.30,random_state=22,stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
Log_reg=LogisticRegression()
```

```python
Log_reg.fit(x_train,y_train)
```

```python
y_test_predict=Log_reg.predict(x_test)
```

```python
y_train_predict=Log_reg.predict(x_train)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,roc_curve,roc_auc_score
```

```python
accuracy=accuracy_score(y_test,y_test_predict)
accuracy
```

```python
con_matr=confusion_matrix(y_test,y_test_predict)
con_matr
```

```python
class_report=classification_report(y_test,y_test_predict)
print(class_report)
```
